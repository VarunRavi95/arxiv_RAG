```python
# Install necessary libraries (if not already installed)
# !pip install pandas

import pandas as pd

# Function to read and clean the CSV
def read_clean_csv(file_path):
    try:
        # Attempt to read with default parameters
        df = pd.read_csv(file_path)
        print("CSV loaded successfully with default parameters.")
        return df
    except pd.errors.ParserError as e:
        print(f"ParserError encountered: {e}")
        print("Attempting to read with engine='python' and on_bad_lines='skip'...")

        try:
            df = pd.read_csv(
                file_path,
                engine='python',
                on_bad_lines='skip'
            )
            print(f"CSV loaded successfully with {len(df)} rows after skipping bad lines.")
            return df
        except Exception as e2:
            print(f"Failed to read CSV with enhanced parameters: {e2}")
            print("Proceeding to read in chunks to isolate issues...")

            # Read in chunks
            chunk_size = 1000
            chunks = []
            problematic_chunks = []
            try:
                for i, chunk in enumerate(pd.read_csv(file_path, chunksize=chunk_size, engine='python', on_bad_lines='skip'), start=1):
                    chunks.append(chunk)
                    print(f"Processed chunk {i}")
                df = pd.concat(chunks, ignore_index=True)
                print(f"CSV loaded successfully with {len(df)} rows from {i} chunks.")
                return df
            except Exception as e3:
                print(f"Failed to read CSV in chunks: {e3}")
                return None

# Replace with your actual file path
csv_file = 'filtered_arxiv_papers_metadata.csv'

# Read and clean the CSV
df = read_clean_csv(csv_file)

# If df is successfully loaded, perform integrity checks
if df is not None:
    print(f"DataFrame shape: {df.shape}")
    print(df.head())
    # Check for missing values in critical columns
    missing_values = df[['arxiv_id', 'title', 'abstract']].isnull().sum()
    print("Missing values in critical columns:")
    print(missing_values)
else:
    print("Failed to load the CSV file.")
```

```python
# Initialize the SentenceTransformer model
from sentence_transformers import SentenceTransformer

embedding_model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
# Function to generate embeddings in batches
def generate_embeddings(model, texts, batch_size=64):
    """
    Generates embeddings for a list of texts using the specified SentenceTransformer model.

    Args:
        model (SentenceTransformer): The embedding model.
        texts (list of str): The texts to embed.
        batch_size (int): Number of texts to process per batch.

    Returns:
        np.ndarray: Array of embeddings.
    """
    embeddings = model.encode(texts, batch_size=batch_size, show_progress_bar=True, convert_to_numpy=True)
    return embeddings
```

```python
# Prepare your texts
texts = df['text'].astype(str).tolist()  # Ensure all entries are strings

# Generate embeddings
document_embeddings = generate_embeddings(embedding_model, texts, batch_size=128)

print(f"Total embeddings generated: {document_embeddings.shape}")
```

```python
import faiss

# Normalize embeddings for cosine similarity
faiss.normalize_L2(document_embeddings)

# Determine the dimensionality
embedding_dimension = document_embeddings.shape[1]

# Initialize FAISS index with Inner Product (for cosine similarity)
index = faiss.IndexFlatIP(embedding_dimension)

# Add embeddings to the index
index.add(document_embeddings)

print(f"FAISS index has {index.ntotal} vectors.")
```

```python
from vertexai.generative_models import GenerativeModel

# Initialize the Gemini Flash model
GEMINI_MODEL_NAME = "gemini-1.5-flash-002"  # Ensure this is the correct model name
gemini_model = GenerativeModel(GEMINI_MODEL_NAME)
```

```python
def retrieve_similar_documents(query, index, df, top_k=5):
    """
    Retrieves the top-K most similar documents to the query.

    Args:
        query (str): The user query.
        index (faiss.Index): The FAISS index.
        df (pd.DataFrame): The DataFrame containing documents.
        top_k (int): Number of top similar documents to retrieve.

    Returns:
        pd.DataFrame: The top-K similar documents.
    """
    # Generate embedding for the query
    query_embedding = embedding_model.encode(query, convert_to_numpy=True)

    # Ensure the embedding is 2D
    if len(query_embedding.shape) == 1:
        query_embedding = query_embedding.reshape(1, -1)

    # Normalize the query embedding
    faiss.normalize_L2(query_embedding)

    # Perform similarity search
    distances, indices = index.search(query_embedding, top_k)

    # Retrieve the corresponding documents
    similar_docs = df.iloc[indices[0]]
    return similar_docs

def create_prompt(query, documents):
    """
    Creates a prompt by combining the user query with retrieved documents.

    Args:
        query (str): The user query.
        documents (pd.DataFrame): The retrieved documents.

    Returns:
        str: The formatted prompt.
    """
    prompt
    prompt = f"Question: {query}\n\n"
    prompt += "Context:\n"
    for idx, row in documents.iterrows():
        prompt += f"- {row['title']}: {row['abstract']}\n"
    prompt += "\nAnswer:"
    return prompt

def generate_response(prompt, gemini_model, temperature=0.2, max_output_tokens=256, top_p=0.8, top_k=40):
    """
    Generates a response using the Gemini Flash model based on the prompt.

    Args:
        prompt (str): The input prompt.
        gemini_model (GenerativeModel): The initialized Gemini Flash model.
        temperature (float): Sampling temperature.
        max_output_tokens (int): Maximum number of tokens to generate.
        top_p (float): Nucleus sampling parameter.
        top_k (int): Top-K sampling parameter.

    Returns:
        str: The generated answer.
    """
    try:
        response = gemini_model.generate_content(
            prompt
        )
        return response.text.strip()
    except Exception as e:
        print(f"Error generating response: {e}")
        return ""
```

```python
while True:
    user_query = input("Enter your question (or type 'exit' to quit): ")
    if user_query.lower() == 'exit':
        print("Exiting the RAG system. Goodbye!")
        break

    # Retrieve similar documents
    similar_docs = retrieve_similar_documents(user_query, index, df, top_k=5)

    if similar_docs.empty:
        print("No similar documents found or an error occurred.")
        continue

    # Create prompt
    prompt = create_prompt(user_query, similar_docs)

    # Generate response using Gemini Flash model
    answer = generate_response(prompt, gemini_model)

    # Display the answer
    print("\nAnswer:", answer, "\n")
```

Approach Using the Google Text Embedding API

```python
def chunked_iterable(iterable, size):
    """Yield successive chunks from iterable of specified size."""
    for i in range(0, len(iterable), size):
        yield iterable[i:i + size]
```

```python
from vertexai.language_models import TextEmbeddingInput, TextEmbeddingModel
import time

def generate_embeddings_with_retry(model, batch, task, retries=3, delay=5):
    inputs = [TextEmbeddingInput(text, task) for text in batch]
    kwargs = dict(output_dimensionality=dimensionality) if dimensionality else {}

    for attempt in range(retries):
        try:
            batch_embeddings = model.get_embeddings(inputs, **kwargs)
            return [embedding.values for embedding in batch_embeddings]
        except Exception as e:
            print(f"Attempt {attempt + 1} failed with error: {e}")
            if attempt < retries - 1:
                print(f"Retrying in {delay} seconds...")
                time.sleep(delay)
            else:
                print("Max retries reached. Skipping this batch.")
                return None
```

```python
EMBEDDING_MODEL = 'text-embedding-004'  # Ensure correct model name
dimensionality = 256
task = "RETRIEVAL_DOCUMENT"
```

```python
model = TextEmbeddingModel.from_pretrained(EMBEDDING_MODEL)

# Prepare your texts
texts = df['text'].astype(str).tolist()  # Ensure all entries are strings

# Initialize a list to hold all embeddings
all_embeddings = []

# Define batch size based on API limits
BATCH_SIZE = 250

# Create an iterator over the batches
batches = list(chunked_iterable(texts, BATCH_SIZE))

# Process each batch
for batch in tqdm(batches, desc="Generating Embeddings"):
    batch_embeddings = generate_embeddings_with_retry(model, batch, task)
    if batch_embeddings:
        all_embeddings.extend(batch_embeddings)
    else:
        print(f"Skipping batch starting with: '{batch[0][:30]}...'")

# Convert the list of embeddings to a NumPy array
document_embeddings = np.array(all_embeddings).astype('float32')

print(f"Total embeddings generated: {document_embeddings.shape}")
```

```python
def get_query_embedding(query_text):
    """
    Generates an embedding for the given query.

    Args:
        query_text (str): The user query.

    Returns:
        np.ndarray: The normalized embedding vector, or None if failed.
    """
    try:
        inputs = [TextEmbeddingInput(query_text, task)]
        kwargs = dict(output_dimensionality=dimensionality) if dimensionality else {}
        response = model.get_embeddings(inputs, **kwargs)
        query_embedding = np.array([embedding.values for embedding in response]).astype('float32')
        faiss.normalize_L2(query_embedding)
        return query_embedding
    except Exception as e:
        print(f"Error generating embedding for query '{query_text}': {e}")
        return None

def retrieve_similar_documents(query, index, df, top_k=5):
    """
    Retrieves the top-K most similar documents to the query.

    Args:
        query (str): The user query.
        index: The FAISS index.
        df (pd.DataFrame): The DataFrame containing documents.
        top_k (int): Number of top similar documents to retrieve.

    Returns:
        pd.DataFrame: The top-K similar documents.
    """
    query_embedding = get_query_embedding(query)
    if query_embedding is None:
        return pd.DataFrame()

    distances, indices = index.search(query_embedding, top_k)
    similar_docs = df.iloc[indices[0]]
    return similar_docs
```
